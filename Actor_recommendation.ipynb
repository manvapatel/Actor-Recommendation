```python
import json
import pandas as pd
import pickle
import keras
from keras.models import Model
from keras.layers import Dense, Input
from keras.utils import to_categorical

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
actor_genre_df = pickle.load( open( "save_genActor.p", "rb" ) )
```

```python
f = lambda x:1/x
```

```python
actor_genre_df["ActorScore"] = actor_genre_df["ActorScore"].apply(f)
actor_genre_df["GenreScore"] = actor_genre_df["GenreScore"].apply(f)
```

```python
actor_genre_df.head()
```

```python
df_gen_score = pd.read_csv('Individulal_Movie_score.csv')
```

```python
df_gen_score.head()
```

```python
actor_genre_df = pd.merge(actor_genre_df,df_gen_score, on = ["Actor", "Genre"])
```

```python
y = actor_genre_df["UserRating"]
```

```python
Y = y
Y.head()
```

```python
X = actor_genre_df.drop(["UserRating","FirstYear","LastYear"], axis = 1)
```

```python
X
```

```python
genre_Dict = {}
actor_Dict = {}
movie_Dict = {}
```

```python
#Converting Genre to integers
genres = X["Genre"]

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(genres)
for i,integer in enumerate(integer_encoded):
    if genres[i] not in genre_Dict:
        genre_Dict[integer] = genres[i]
    

X["Genre"] = integer_encoded

#Converting actors to integers
actors = X["Actor"]

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(actors)

for i,integer in enumerate(integer_encoded):
    if actors[i] not in actor_Dict:
        actor_Dict[integer] = actors[i]
        
X["Actor"] = integer_encoded

movie = X["Title"]

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(movie)

for i,integer in enumerate(integer_encoded):
    if movie[i] not in movie_Dict:
        movie_Dict[integer] = movie[i]
        
X["Title"] = integer_encoded

```

```python
actor_Dict
```

```python
X
```

```python
#Normalizing

cols_to_normalize = ["Votes", "Gross", "NumberofMovies", "ShortMovie", "LongMovie"]

x = X[cols_to_normalize].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X[cols_to_normalize] = pd.DataFrame(x_scaled)
```

```python
X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=9000)
```

```python
y_train.shape
```

```python
x_train.shape[1:]
```

```python
input_layer = Input(shape=x_train.shape[1:])
m = Dense(x_train.shape[1], activation='relu')(input_layer)
m = Dense(8, activation='relu')(m)
m = Dense(1)(m)
```

```python
model = Model (inputs = input_layer, outputs = m)
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
history = model.fit(x_train, y_train,
                    epochs=3,
                    batch_size=32,
                    validation_split=0.1,
                    verbose = 1)
```

```python
predictions = model.predict(x_test)
```
